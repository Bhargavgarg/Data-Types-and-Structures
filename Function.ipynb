1. What is the difference between a function and a method in Python?

 Ans>>Function: A block of reusable code defined with def or lambda, not tied to any object.

Method: A function that is associated with an object (usually defined inside a class) and operates on that object.

Example:
# Function

    return f"Hello {name}"

# Method
class Person:
    def greet(self):

        return "Hello from method!"

print(p.greet())

6. Explain the concept of generators in Python and how they are defined.

Ans>>
A generator is a function that uses yield to return values one at a time, instead of returning them all at once.

It produces values lazily (on demand).

Example:

def count_up_to(n):

    for i in range(1, n+1):
        yield i

for num in count_up_to(5):

    print(num)

7. What are the advantages of using generators over regular functions?

Ans>>
#Advantages of Generators

Memory efficient (no need to store whole list in memory).

Lazy evaluation (values produced when needed).

Useful for infinite sequences.

Example:

    for i in range(n):

        yield i

print(next(gen))

10. What is the difference between `map()`, `reduce()`, and `filter()`
functions in Python?

Ans>>map() → Applies a function to each element and returns new iterable.

reduce() → Reduces iterable to a single value by repeatedly applying function.

filter() → Returns elements that satisfy a condition.

Example:
from functools import reduce

nums = [1, 2, 3, 4, 5]

print(list(map(lambda x: x*2, nums)))      

print(reduce(lambda a,b: a+b, nums))       

print(list(filter(lambda x: x%2==0, nums)))

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))   # True
print(is_prime(20))  # False
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms.
class Fibonacci:
    def __init__(self, limit):
        self.limit = limit
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.limit:
            raise StopIteration
        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

fib = Fibonacci(10)
print(list(fib))  # first 10 Fibonacci numbers
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i

print(list(powers_of_two(5)))  # [1, 2, 4, 8, 16, 32]
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
tuples_list = [(1, 5), (2, 3), (4, 1)]
sorted_list = sorted(tuples_list, key=lambda x: x[1])
print(sorted_list)  # [(4, 1), (2, 3), (1, 5)]
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)  # [32.0, 68.0, 98.6, 212.0]
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    return ''.join(filter(lambda ch: ch.lower() not in "aeiou", s))

print(remove_vowels("Hello World"))  # Hll Wrld
```

```python
# 11.Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

# Order Number    Book Title and Author                   Quantity    Price per Item
# 34587           Learning Python, Mark Lutz              4           40.95
# 98762           Programming Python, Mark Lutz           5           56.80
# 77226           Head First Python, Paul Barry           3           32.95
# 88112           Einführung in Python3, Bernd Klein      3           24.99


# Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
# product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
# order is smaller than 100,00 €.

#Write a Python program using lambda and map.
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

result = list(map(
    lambda order: (order[0], order[2]*order[3] if order[2]*order[3] > 100 else order[2]*order[3] + 10),
    orders
))

print(result)
# [(34587, 163.8), (98762, 284.0), (77226, 108.85), (88112, 84.97)]
```
